```python
def pattern_match(text):
    matcher = Matcher(nlp.vocab)
    # Add match ID "HypothesisIs" with no callback and one pattern
    pattern = [{'LEMMA':{"IN":["hypothesis","hypothesize","hypothesise", "hypothesized", "hypothesised"]}}]
#     pattern1 = [{'POS':'PRON','OP':'*'},{'POS': 'AUX', 'OP':'*'},{'LEMMA':{"IN":["hypothesis","hypothesize","hypothesise"]}},{'POS':'SCONJ', 'OP':'*'},{'POS':'NOUN','OP':'*'}]
#     pattern2 = [{'POS':'PRON','OP':'*'},{'LEMMA':{"IN":["hypothesis","hypothesize","hypothesise"]}},{'POS':'SCONJ', 'OP':'*'},{'POS':'NOUN','OP':'*'}]
#     pattern3 = [{'POS':'NOUN','OP':'*'},{'LEMMA':{"IN":["hypothesis","hypothesize","hypothesise"]}},{'POS':'SCONJ', 'OP':'*'},{'POS':'NOUN','OP':'*'}]
    
    matcher.add("HypothesisIs", None, pattern)

    doc = nlp(text)
    matches = matcher(doc)
    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]  # Get string representation
        span = doc[start:end]  # The matched span
        sent = span.sent
        return sent.text
```

```python
# Pattern match is run on the 'object' column
data_df['pattern_match'] = data_df['object'].apply(pattern_match)
```

## Finding unique patterns

```python
# Visualizing the sentence fragments that match:

matcher = Matcher(nlp.vocab)
matched_sents = []  # Collect data of matched sentences to be visualized

def collect_sents(matcher, doc, i, matches):
    match_id, start, end = matches[i]
    span = doc[start:end]  # Matched span
    sent = span.sent  # Sentence containing matched span
    # Append mock entity for match in displaCy style to matched_sents
    # get the match span by ofsetting the start and end of the span with the
    # start and end of the sentence in the doc
    match_ents = [{
        "start": span.start_char - sent.start_char,
        "end": span.end_char - sent.start_char,
        "label": "",
    }]
    matched_sents.append({"text": sent.text, "ents": match_ents})

pattern1 = [{'POS':'PRON','OP':'+'},{'POS': 'AUX', 'OP':'+'},{'LEMMA':{"IN":["hypothesis","hypothesize","hypothesise"]}},{'POS':'SCONJ', 'OP':'+'},{'POS':'NOUN','OP':'+'}]
pattern2 = [{'POS':'PRON','OP':'+'},{'LEMMA':{"IN":["hypothesis","hypothesize","hypothesise"]}},{'POS':'SCONJ', 'OP':'+'},{'POS':'NOUN','OP':'+'}]
pattern3 = [{'POS':'DET','OP':'+'},{'LEMMA':{"IN":["hypothesis","hypothesize","hypothesise"]}},{'POS':'SCONJ', 'OP':'+'},{'POS':'NOUN','OP':'+'}]
matcher.add("HypothesisIs", collect_sents, pattern1, pattern2, pattern3)  # add pattern
doc = nlp(doc_variable)
matches = matcher(doc)

# Serve visualization of sentences containing match with displaCy
# set manual=True to make displaCy render straight from a dictionary
# (if you're not running the code within a Jupyer environment, you can
# use displacy.serve instead)
displacy.render(matched_sents, style="ent", manual=True)
```

```python
# Understanding parts of speech
doc = nlp(u'Based on the data in hand, we hypothesize that sepsis induces imbalances in microbial populations residing in the gut, along with compromises in epithelial integrity.')
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
    
# Text: The original word text.
# Lemma: The base form of the word.
# POS: The simple UPOS part-of-speech tag.
# Tag: The detailed part-of-speech tag.
# Dep: Syntactic dependency, i.e. the relation between tokens.
# Shape: The word shape – capitalization, punctuation, digits.
# is alpha: Is the token an alpha character?
# is stop: Is the token part of a stop list, i.e. the most common words of the language?

# https://spacy.io/usage/linguistic-features/
```

```python
doc = nlp(u'Based on the data in hand, we hypothesize that sepsis induces imbalances in microbial populations residing in the gut, along with compromises in epithelial integrity.')
for chunk in doc.noun_chunks:
    print(chunk.text, chunk.root.text, chunk.root.dep_,
            chunk.root.head.text)
    
# Text: The original noun chunk text.
# Root text: The original text of the word connecting the noun chunk to the rest of the parse.
# Root dep: Dependency relation connecting the root to its head.
# Root head text: The text of the root token’s head.
```
